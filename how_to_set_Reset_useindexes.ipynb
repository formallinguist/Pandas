```python
#We can use dictionary of lists to create a data frame.
info_people = {
    "first_name":["Ravi","Teja","Divya","Naveena"],
    "last_name": ["chikkala","Theetla","Pingali","Yanam"],
    "email":["ravichikkala@mail.com","Tejatheetla@mail.com","Divyapingalie@mail.com","Naveenayanam@MAIL.COM"]
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(info_people)
```

```python
df
```

```python
df['email']
```

```python
#Setting email as the index for the dataframe.
df.set_index('email')
```

```python
df #Here the data frame has not changed.
```

```python
df.set_index('email',inplace=True) #Now df has changed its data
```

```python
df
```

```python
df.index
```

```python
#getting the row information for that specific index.
df.loc["Divyapingalie@mail.com"]
```

```python
df.loc["Divyapingalie@mail.com","last_name"]
```

```python
df.iloc[0]
```

```python
#getting the df back to normal.
df.reset_index(inplace=True)
```

```python
df #we see that df is changed back to the original df.
```
