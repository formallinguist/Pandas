```python
#We can use dictionary of lists to create a data frame.
info_people = {
    "first_name":["Ravi","Teja","Divya","Naveena"],
    "last_name": ["chikkala","Theetla","Pingali","Yanam"],
    "email":["ravichikkala@mail.com","Tejatheetla@mail.com","Divyapingalie@mail.com","Naveenayanam@MAIL.COM"]
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(info_people)
```

```python
df
```

```python
#using filter criteria on df.
df['last_name'] == 'Pingali'
```

```python
fil_ter = (df['last_name'] == 'Pingali')
```

```python
df[fil_ter]
```

```python
df.loc[fil_ter]
```

```python
df.loc[fil_ter,'email']
```

```python
#Using "and" operator for the forst name and last name
fil_ter = (df['last_name'] == 'Pingali') & ( df['first_name'] == "Divya")
```

```python
df.loc[fil_ter,'email']
```

```python
#Using "or" operator for the forst name and last name
fil_ter = (df['last_name'] == 'Pingali') | ( df['first_name'] == "Ravi")
```

```python
df.loc[fil_ter,'email']
```
