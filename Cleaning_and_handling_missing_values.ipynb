```python
import pandas as pd
import numpy as np
```

```python
people = {
        'first': ['Rohan','Daisy','maumita','satrupa',np.nan,None,'NA'],
        'last': ['Ahmed','boro','bahumik','sarkar',np.nan,np.nan,'Missing'],
        'email':['Rohanhasan@gmail.com','daisyboro@gmail.com','satrupasarkar@gmail.com',np.nan,np.nan,'Missing','Missing'],
        'age': ['33','54','56','28','34','46','67']
    
}
```

```python
df = pd.DataFrame(people)

df.replace("NA",np.nan,inplace=True)
df.replace("Missing",np.nan,inplace=True)
```

```python
df
```

```python
df.dropna() #It drops the na values.
```

```python
df.dropna(axis='index',how = 'all')
```

```python
df.dropna(axis='columns',how = 'all')
```

```python
df.dropna(axis='index',how = 'any',subset= ['email']) #Looking at email address for missing values
```

```python
df.dropna(axis='index',how = 'any',subset= ['last','email']) #Looking at email address for missing values
```

```python
df.isna()
```

```python
df.fillna(0) #replacing na values with zero
```

```python
df.dtypes
```

```python
type(np.nan)
```

```python
df['age'] = df['age'].astype(float)
```

```python
df['age'].mean()
```
