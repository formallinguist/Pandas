# Here we will see how apply function works on series object.

```python
df['EMAIL'].apply(len) # It gives the  length og the email address of each row.
```

```python
def update_email(email):
    return email.upper()
```

```python
df['EMAIL'].apply(update_email) # we are passing the function not execting the function.
```

```python
df['EMAIL'] = df['EMAIL'].apply(update_email)
```

```python
df
```

```python
#Using the lambda function on the dataframe with apply.
df['EMAIL'] = df['EMAIL'].apply(lambda x: x.lower()) #Here we converted the mails to lower case.
```

```python
df
```

```python
df['EMAIL'].apply(len) # Here we are accesing EMAIL coloumn and are getting the length of it.
```

```python
df.apply(len) # Here it is checking each coloumn. or it is applying the length function to each series of the coloumn.
```

```python
df.apply(len,axis='columns') # Here it is counting the legth of the column.
```
