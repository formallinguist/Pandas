#Finding out what percentage of people in a country use python.

```python
country_respon = df['Country'].value_counts()
country_respon 
```

```python
Country_uses_python = country_group["LanguageHaveWorkedWith"].apply(lambda x: x.str.contains('Python').sum())
Country_uses_python
```

```python
df_python = pd.concat([Country_uses_python,country_respon],axis="columns",sort=False)
```

```python
df_python
```

```python
#Renaming coloumns.
df_python.rename(columns={"Country":"NumRespondents","LanguageHaveWorkedWith":"NumKnowsPython"})
```

```python
df_python
```

```python
df_python["Percentage_knows_python"] = (df_python["count"]/df_python["LanguageHaveWorkedWith"]) * 100
```

```python
df_python
```

```python
df_python.sort_values(by="Percentage_knows_python",ascending=False,inplace=True)
```
